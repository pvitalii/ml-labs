```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fetch_data(file_name):
    X = np.array([])
    Y = np.array([])
    with open(file_name) as f:
        data = f.readlines()
        data.pop(0)
        for line in data:
            index,x,y = line.strip().split(",")
            X = np.append(X, float(x))
            Y = np.append(Y, float(y))
    return X, Y

train_data = fetch_data("lab_1_train.csv")
plt.scatter(train_data[0], train_data[1])
plt.show()
```

```python
def gradient_descent(X, Y): 
    b = np.ones(2)
    LR = 0.1  
    epochs = 1000 
    n = X.shape[0] 

    for i in range(epochs): 
        Y_pred = b[0]*X + b[1]
        MSE = (1 / n) * sum((Y_pred - Y) ** 2)
        D_b0 = (-2/n) * sum(X * (Y - Y_pred))
        D_b1 = (-2/n) * sum(Y - Y_pred)
        b[0] = b[0] - LR * D_b0  
        b[1] = b[1] - LR * D_b1
        print(f"epoch {i}\ncost: {MSE}\nb0: {b[0]}, b1: {b[1]}\n")
    return Y_pred

train_gradient = gradient_descent(train_data[0], train_data[1])
```

```python
test_data = fetch_data("lab_1_test.csv")
gradient_descent(test_data[0], test_data[1])
```
